# Principal Component Analysis
Dataset:  
The Yale Face Database B consists of single light source images of $10$ subjects, each seen in different poses and illumination conditions. In the provided form the database is divided into $5$ subsets. In subset $0$ the subject is illuminated by an almost frontal light source, while for subsets $1$-$4$ the light source is gradually moved along the horizon. Subset $0$ will serve as the training set, while subsets $1$-$4$ are used for testing.

## 1. PCA 
**Aplly PCA to reduce the dimensionality of the data while preserving as much information as possible. Visualize the eigenface of first PC.**  

Write a function that takes as an input matrix $\mathbf{T}$ of vectorized images from subset $0$. The output of this function are the $20$ first singular vectors $\mathbf{U}[:, 1]\,,\;...\,,\;\mathbf{U}[:, 20]$. Display the first $3$ vectors as images, i.e., reshape them to size $50 \times 50$ and display them.

```python
import numpy as np
import imageio
from os import listdir
from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsClassifier as KNN
```

```python
subset_nums = [0, 1, 2, 3, 4]                                   # Load all the subsets of data
img_width = 50
img_height = 50
max_uint8 = 2**8 - 1
directory = 'yaleBfaces'
subsets = {}
labels = {}

for num in subset_nums:
    subset_name = 'subset' + str(num)
    filenames = listdir(directory + '/' + subset_name)          # Get all the filenames in the folder
    num_files = len(filenames)                                  # Initialize array
    subsets[num] = np.zeros((img_width * img_height, num_files))
    labels[num] = np.zeros(num_files)
    samp = 0                                                    # Load all files into particular subset
    for f_name in filenames:
        curr_f_name = directory + '/' + subset_name + '/' + f_name
        subsets[num][:, samp] = np.reshape(np.array(imageio.imread(curr_f_name)) / max_uint8, (img_width * img_height))
        labels[num][samp] = int(f_name[6:8])
        samp += 1
```

```python
def get_pcs(M, n):                                              # Input matrix M and returns 20 first singular vectors
    m = np.mean(M, axis = 1).reshape(M.shape[0], 1)             # Centre M (p Ã— N matrix)
    M_c = M - m
    U, Sigma, VT = np.linalg.svd(M_c, full_matrices = False)    # Take SVD of M_c
    
    return U[:, 0:n]                                            # Return first n singular vectors

M = subsets[0]
n = 20
U = get_pcs(M, n)
```

```python
Un_visual = {}                                                               # Visualize the first three PC vectors
num_vecs = 3

for col in range(0, num_vecs):                                               # Reshape columns of Un
    Un_visual[col] = U[:, col].reshape(img_height, img_width)

fig, ax = plt.subplots(ncols = num_vecs, sharey = True, figsize = (10,3))    # Plot resulting images
for col in range(0, num_vecs):
    ax[col].set_title('Vector ' + str(col + 1))
    ax[col].imshow(Un_visual[col][:, :])
    ax[col].set_xlabel('x')
    ax[col].set_ylabel('y')
 
plt.show()
```

### 2. KNN
**In the space of lower dimensionality after applying PCA, use the Euclidean distance to classify each sample image based on its three nearest neighbors. Report the error rate.**  

Write a function that takes as an input the training set $\mathbf{T}$ (a matrix composed of vectorized pictures from subset $0$), a vector containing the labels of the training set (i.e., if the the $i$-th sample belongs to class $j$, the $i$-th entry of the labels vector is $j$. In this exercise $j$ is an integer between $1$ and $10$), the test samples $S$ (a matrix composed of vectorized pictures from subsets $1$-$4$) and the corresponding labels (in a separate vector), the $20$ singular vectors from the first step, and the parameter $k$ that denotes how many of the PCs are used. Use the Euclidean distance to classify each sample image based on its three nearest neighbors (This is done by comparing the test samples with the training samples in the reduced space.). As an output give the fraction of images from $S$ that were misclassified, i.e., the error rate. Repeat this for subsets $1$ through $4$ and for $k=1\,,\;...\,,\;20$. Plot the error rate for each subset.

```python
def calc_error_rate(M, M_labs, S, S_labs, U, k):      # Calculate the error rate
    # M/S is p Ã— N matrix (N varying according to samples in subset)
    # M_labs / S_labs is size N vector
    # U is p Ã— k matrix
    # k is scalar value    
    m = np.mean(M, axis = 1).reshape(M.shape[0], 1)
    s = np.mean(S, axis = 1).reshape(S.shape[0], 1)
    
    model = KNN(n_neighbors = 3)
    Uk = U[:, 0:k]
    UkM = np.transpose(Uk)                            # PC transformation for dimensionality reduction
    M_new = np.dot(UkM, M-m)
    S_new = np.dot(UkM, S-s)
    model.fit(np.transpose(M_new), M_labs)            # Train KNN on M_new
    acc = model.score(np.transpose(S_new), S_labs)    # Test KNN on S
    
    return 1 - acc                                    # Output fraction of the misclassified samples
```

```python
k_nums = range(1, 21)                                 # Calculate errors of each subset and k value
err = {}
M = subsets[0]
M_labs = labels[0]

for num in subset_nums[1:]:                           # Loop through all subsets

    err[num] = np.zeros(len(k_nums))
    index_counter = 0                                 # Calculate error as k increases
    for k in k_nums:
        S = subsets[num]
        S_labs = labels[num]
        err[num][index_counter] = calc_error_rate(M, M_labs, S, S_labs, U, k)
        index_counter += 1
```

```python
for num in subset_nums[1:]:                           # Visualize the error rates
    plt.plot(k_nums, err[num], label='subset ' + str(num))

plt.title('Error Rates of Subsets 1-4')
plt.xlabel('k')
plt.ylabel('Error Rates')
plt.grid()
plt.legend()
plt.show()
```

Repeat the above experiment without using the first three singular vectors, i.e., use $k=1\,,\;...\,,\;17$ singular vectors starting from the $4$-th. Plot the error rate as before. How do you explain the difference in recognition rate?

```python
U_r = U[:, 3:]                                                                # Remove first three columns of k
k_nums_r = range(1, 18)                                                       # Calculate errors of each subset and k
err_r = {}
M = subsets[0]
M_labs = labels[0]

for num in subset_nums[1:]:                                                   # Loop through all subsets
    err_r[num] = np.zeros(len(k_nums_r))
    index_counter = 0                                                         # Calculate error as k increases
    for k in k_nums_r:
        S = subsets[num]
        S_labs = labels[num]
        err_r[num][index_counter] = calc_error_rate(M, M_labs, S, S_labs, U_r, k)
        index_counter += 1


fig, ax = plt.subplots(len(subset_nums[1:]), sharex=True, figsize=(10,12))    # Visualize the error rates
subplt_index = 0
for num in subset_nums[1:]:
    ax[subplt_index].plot(k_nums, err[num], color='red', label='No columns of $U_k$ removed')
    ax[subplt_index].plot(k_nums_r, err_r[num], color='blue', label='First 3 columns of $U_k$ removed')
    ax[subplt_index].set_title('Error Rate of Subset ' + str(num))
    ax[subplt_index].set_xlabel('k')
    ax[subplt_index].set_ylabel('Error Rate')
    ax[subplt_index].grid()
    ax[subplt_index].legend()

    subplt_index += 1
    
plt.show()
```

** Solution:**

For subset 1: the first 3 columns of $U_k$ does not matter so much. This may because the other PCs have similar ability to separate samples by labels like the first 3 PCs do.

For subsets 2 & 3: without the first 3 columns of $U_k$ leads to better classification performance. This may because the other PCs are better at separating samples by labels, allowing KNN to correctly classify samples, although the first 3 PCs represent more variance of the samples.

For subset 4: removing first 3 columns of $U_k$ causes worse performance. This may because the first 3 PCs are better at separating samples by labels compared to the other PCs, however both errors remained quite high and close.

Conclusion: Classification ability is dependent on the separating-samples-by-labels ability of PCs.
