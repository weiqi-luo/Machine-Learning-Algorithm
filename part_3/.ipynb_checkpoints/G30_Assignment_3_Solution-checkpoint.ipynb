### The Kernel Trick
### Task 1:
**On Moodle you will find a Jupyter-Notebook that contains a function for dimensionality reduction via PCA. The function `linear_pca` expects a data matrix $\mathbf{X}\in \mathbb{R}^{p\times N}$ and a number of PCs $k$ and returns the first $k$ PCA scores for the matrix $\mathbf{X}$.**
- **Provide code that tests the function with selected images from the provided MNIST training dataset by visualizing the first 2 scores in a scatter plot.**

```python
import numpy as np
import imageio
from os import listdir
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.patches as mpatches
%matplotlib inline
```

```python
def linear_pca(X, k):
    mu=np.mean(X,axis=1)
    X_c=X-np.expand_dims(mu,axis=1)
    [U,Sigma,VT]=np.linalg.svd(X_c,full_matrices=False)
    S=np.expand_dims(Sigma[:k],axis=1)*VT[:k,:]
    return S
```

```python
# Insert Code for Subtask 1.1 here
directory = 'mnist'                                      # Set number of classes
num_classes = len(listdir(directory))
digit_pref=['d%d' % i for i in range(0, num_classes-1)]

colors = cm.rainbow(np.linspace(0, 1, num_classes))      # Sample colors from spectrum to represent each class

n = np.zeros(num_classes)                                # Get the number of samples in each class
for i, dp in enumerate(digit_pref):
    class_samples = listdir(directory + '/' + dp)
    n[i] = 100                                           # Number of samples to read from each class

N = int(sum(n))                                          # Total samples
X = np.zeros((784,N))                                    # Data matrix
Y = np.zeros(N)                                          # Label vector
C = np.zeros((N, 4))                                     # Color vector

for i, dp in enumerate(digit_pref):                      # Read data and populate X, Y and C
    Y[int(sum(n[:i])):int(sum(n[:i+1]))] = i
    C[int(sum(n[:i])):int(sum(n[:i+1])), :] = np.expand_dims(colors[i],axis=0)
    for j in range(int(n[i])):X[:, int(sum(n[:i]))+j] = np.float64(imageio.imread('mnist/'+dp+'/'+dp+'_'+'%04d.png'%(j+1)).ravel())
        
k = 2                                                    # Get first k PCs of data
S = linear_pca(X, k)

idx_new = np.arange(N)                                   # Randomize order of data for plotting
np.random.shuffle(idx_new)

PC1 = S[0, idx_new]                                      # First 2 PCs
PC2 = S[1, idx_new]

plt.scatter(PC1, PC2, c=C[idx_new, :], s=20)             # Plot the first 2 PCs of data
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.title('MNIST PCs via Function: linear_pca')
legend_patch = [mpatches.Patch(color=colors[i], label=str(i)) for i in range(0, num_classes)]
plt.legend(handles=legend_patch)
plt.grid()
plt.show()
```

- **Complete the function `gram_pca` such that it has the same functionality as `linear_pca` but expects a gram matrix $\mathbf{K}=\mathbf{X}^\mathrm{T}\mathbf{X}$ instead of the data matrix $\mathbf{X}$ as its input. Do not assume that $\mathbf{K}$ was produced from centered data. Note: It is important to be consistent in notation here. E.g., for a data matrix of $1000$ MNIST images, we have $\mathbf{X}\in \mathbb{R}^{784 \times 1000}$ and $\mathbf{K}\in \mathbb{R}^{1000 \times 1000}$.**

```python
def gram_pca(K,k):
    # Insert Code for Subtask 1.2 here
    N = len(K)                                                   # Dimension of K
    I = np.eye(N)                                                # NN identity matrix
    A = np.full((N, N), 1./N)                                    # NN matrix with values 1/N
    K_c = (I - A) @ K @ (I - A)                                  # Centered gram matrix
    [V, Sigma2, VT] = np.linalg.svd(K_c, full_matrices = False)  # SVD of K_c and PCs
    Sigma = np.sqrt(Sigma2)
    S = np.expand_dims(Sigma[:k], axis = 1) * VT[:k, :]
    return S
```

- **There is an unknown vector space $\mathbb{H}$, equipped with an inner product $\langle \cdot, \cdot \rangle_{\mathbb{H}}$ and a function**

$$
\varphi : \mathbb{R}^p \rightarrow \mathbb{H},
$$

**such that**

$$
\langle \varphi(\mathbf{x}) , \varphi(\mathbf{y}) \rangle_\mathbb{H} = \exp{(-\frac{{\|\mathbf{x}-\mathbf{y}\|}^2}{2\sigma^2})}
$$
**holds for every $\mathbf{x}, \mathbf{y} \in \mathbb{R}^p$. The expression on the right-hand side of the equation is called the *Gaussian kernel* and $\sigma$ is a parameter to choose by hand.**

**The function `gaussian_kernel_pca` expects a data matrix $\mathbf{X}$, a reduced dimension number $k$ and a parameter $\sigma$. It returns the first $k$ *Kernel PCA* scores of the data. In other words, the function returns the first $k$ PCA scores of**

$$
\varphi(\mathbf{x}_1), \varphi(\mathbf{x}_2), ..., \varphi(\mathbf{x}_N),
$$

**where $\mathbf{x}_i$ denotes the $i$-th data sample/$i$-th column of the data matrix. The function `gaussian_kernel_pca` is already written, but for it to work, the function `compute_gaussian_gram_matrix` must return correct results. Complete `compute_gaussian_gram_matrix` accordingly.**

```python
def compute_gaussian_gram_matrix(X, sigma):
    #Insert Code for Subtask 1.4 here
    K = np.zeros((X.shape[1], X.shape[1]))    
    for i in range(X.shape[1]):
        for j in range(i):                             # Only compute K above main diagonal inclusive            
            pair_diff = X[:, i] - X[:, j]              # Get all pair-wise differences in data            
            sq_norm = np.sum(pair_diff*pair_diff)      # Take the squared norm for each of them            
            K[i, j] = np.exp(-sq_norm/(2*(sigma**2)))  # Plug each squared norm into the gaussian kernel            
            K[j, i] = K[i, j]                          # K is symmetric
            
    return K

def gaussian_kernel_pca(X, k, sigma):
    K=compute_gaussian_gram_matrix(X,sigma)
    S=gram_pca(K,k)
    return S
```

- **Test `gaussian_kernel_pca` with some MNIST train images and $\sigma = 1000$.**

```python
#Insert Code for Subtask 1.5 here
k = 2                                                     # Get first k PCs of data via gaussian_kernel_pca
sigma = 1000
S_gauss = gaussian_kernel_pca(X, k, sigma)

PC1_gauss = S_gauss[0, idx_new]                           # First 2 PCs via gaussian_kernel_pca
PC2_gauss = S_gauss[1, idx_new]

plt.scatter(PC1_gauss, PC2_gauss, c=C[idx_new, :], s=20)  # Plot the first 2 gaussian-kernel PCs of data
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.title('MNIST PCs via Function: gaussian_kernel_pca')
legend_patch = [mpatches.Patch(color=colors[i], label=str(i)) for i in range(0, num_classes)]
plt.legend(handles=legend_patch)
plt.grid()
plt.show()
```
